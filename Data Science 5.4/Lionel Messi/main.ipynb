# Lionel Messi

import libraries :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

read data and show the head of data :

```python
data = pd.read_csv("Lionel Messi Goals.csv")
data.head()
```

```python
def year(date):
    new_str = str(date)
    return new_str[-4:-1]+new_str[-1]
```

```python
data["Date"] = data["Date"].apply(year)
df = data["Date"].value_counts().reset_index()
df = df.sort_values(by=["Date"])
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["Date"],df["count"],color="red",marker='o')
plt.title("The number of goals scored by Lionel Messi each year")
plt.xlabel("Year")
plt.ylabel("number of goals")
plt.show()
```

Processing the types of goals scored by Lionel Messi and displaying them in a pie :

```python
df = data.query("`Goal Type` == 'Penalty'")
df = df["Goal Method"].value_counts().reset_index()
df
```

```python
plt.pie(df["count"],labels=df["Goal Method"],autopct='%1.1f%%')
plt.title("Lionel Messi's penalty goal foot type")
plt.show()
```
