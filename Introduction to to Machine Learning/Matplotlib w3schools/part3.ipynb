```python
import matplotlib.pyplot as plt

# داده‌ها
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 10]
y3 = [100, 200, 300, 400]
# ایجاد نمودار اول
plt.subplot(1, 3, 1)  # 1 سطر، 2 ستون، نمودار اول
plt.plot(x, y1)
plt.title("Line 1")

# ایجاد نمودار دوم
plt.subplot(1, 3, 2)  # 1 سطر، 2 ستون، نمودار دوم
plt.plot(x, y2)
plt.title("Line 2")


plt.subplot(1,3,3)  # 1 سطر، 3 ستون، نمودار دوم
plt.plot(x, y3)
plt.title("Line 3")

# نمایش نمودارها
plt.show()
```

```python
import matplotlib.pyplot as plt

# داده‌ها
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# رسم نمودار با سبک‌های مختلف
plt.plot(x, y, color='red', linestyle='--', marker='o')

# نمایش نمودار
plt.show()
```
