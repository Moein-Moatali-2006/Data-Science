```python
import matplotlib.pyplot as plt

# Subplot اول
plt.subplot(2, 2, 1)
plt.plot([0, 1, 2], [10, 20, 30])
plt.title("Plot 1")

# Subplot دوم
plt.subplot(2, 2, 2)
plt.plot([0, 1, 2], [30, 20, 10])
plt.title("Plot 2")

# Subplot سوم
plt.subplot(2, 2, 3)
plt.plot([0, 1, 2], [10, 15, 10])
plt.title("Plot 3")

# Subplot چهارم
plt.subplot(2, 2, 4)
plt.plot([0, 1, 2], [15, 10, 15])
plt.title("Plot 4")

# نمایش
plt.tight_layout()
plt.show()
```
