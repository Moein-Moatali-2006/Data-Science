```python
import matplotlib.pyplot as plt

# داده‌ها
values = [30, 20, 40, 10]
labels = ["Category 1", "Category 2", "Category 3", "Category 4"]
explode = [0, 0.2, 0.3, 0]  # بخش سوم برجسته شود

# رسم نمودار دایره‌ای
plt.pie(values, labels=labels, autopct="%1.1f%%", explode=explode)

# نمایش
plt.show()
```
