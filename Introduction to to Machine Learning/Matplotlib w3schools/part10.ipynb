```python
import matplotlib.pyplot as plt

# داده‌ها
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
sizes = [20, 50, 100, 200, 500, 1000, 60, 90, 30, 300]
colors = [20, 50, 100, 200, 250, 300, 400, 500, 600, 700]

# رسم نمودار پراکندگی با رنگ‌ها و اندازه‌های متفاوت
plt.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.7)

# اضافه کردن نوار رنگی (Colorbar)
plt.colorbar()

# نمایش
plt.show()
```

```python
import matplotlib.pyplot as plt

# داده‌های مجموعه اول
x1 = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y1 = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# داده‌های مجموعه دوم
x2 = [2, 2, 8, 10, 6, 4, 5, 8, 6, 5]
y2 = [100, 105, 84, 90, 92, 99, 102, 110, 85, 88]

# رسم نمودار پراکندگی
plt.scatter(x1, y1, color="red", label="Group 1", alpha=0.7)
plt.scatter(x2, y2, color="blue", label="Group 2", alpha=0.7)

# اضافه کردن برچسب و عنوان
plt.title("Scatter Plot with Multiple Groups")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()

# نمایش
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
