```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# رسم نمودار
plt.plot(x, y)

plt.title("Hi Matplotlib")
plt.xlabel("This is X")
plt.ylabel("This is Y")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

# رسم نمودار
plt.plot(x, y)

plt.title("Hi Matplotlib")
plt.xlabel("This is X")
plt.ylabel("This is Y")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

# رسم نمودار
plt.plot(x, y1,label="this is Line 1")
plt.plot(x, y2,label="this is Line 2")

plt.legend()

plt.savefig("plt_1.png")
plt.show()
```
