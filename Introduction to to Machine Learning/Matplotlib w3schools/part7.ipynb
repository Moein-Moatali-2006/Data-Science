```python
import matplotlib.pyplot as plt

# داده‌ها
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# رسم نمودار
plt.plot(x, y)

# اضافه کردن برچسب‌ها و عنوان با فرمول ریاضی
plt.xlabel("X Label ($x^2 + y^2 = z^2$)")  # فرمول ریاضی در برچسب
plt.ylabel("Y Label ($\\alpha + \\beta = \\gamma$)")
plt.title("Math Example: $E=mc^2$")  # عنوان با فرمول ریاضی

# نمایش نمودار
plt.show()
```
