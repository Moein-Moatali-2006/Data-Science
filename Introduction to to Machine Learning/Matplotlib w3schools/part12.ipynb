```python
import matplotlib.pyplot as plt

# داده‌ها
data1 = [1, 2, 2, 3, 3, 3, 4, 4, 4, 5]
data2 = [3, 4, 4, 5, 5, 5, 6, 6, 6, 7]

# رسم هیستوگرام‌ها
plt.hist(data1, bins=5, alpha=0.5, label="Dataset 1", color="blue", edgecolor="black")
plt.hist(data2, bins=5, alpha=0.5, label="Dataset 2", color="green", edgecolor="black")

# اضافه کردن برچسب‌ها و عنوان
plt.title("Comparison of Two Histograms")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.legend()

# نمایش
plt.show()
```
