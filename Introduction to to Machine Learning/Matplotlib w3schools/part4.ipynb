```python
import matplotlib.pyplot as plt

# داده‌ها
x = ['A', 'B', 'C', 'D']
y = [10, 20, 30, 40]

# رسم نمودار میله‌ای
plt.bar(x, y)

# نمایش نمودار
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# داده‌ها
x = [random.randint(1,1000), random.randint(1,1000), random.randint(1,1000), random.randint(1,1000), random.randint(1,1000)]
y = [random.randint(1,1000), random.randint(1,1000), random.randint(1,1000), random.randint(1,1000), random.randint(1,1000)]

# رسم نمودار پراکندگی
plt.scatter(x, y)

# نمایش نمودار
plt.show()
```

```python
import matplotlib.pyplot as plt

# داده‌ها
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

# رسم نمودار دایره‌ای
plt.pie(sizes, labels=labels,autopct='%1.1f%%')

# نمایش نمودار
plt.show()
```

```python
import matplotlib.pyplot as plt

# داده‌ها
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
yerr = [1, 2, 1.5, 2.5]  # نوار خطا

# رسم نمودار با نوار خطا
plt.errorbar(x, y, yerr=yerr, fmt='o')

# نمایش نمودار
plt.show()
```
