```python
import matplotlib.pyplot as plt

# داده‌ها
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# رسم نمودار
plt.plot(x, y)

# اضافه کردن خطوط شبکه فقط به محور افقی
plt.grid(axis='x')

# نمایش نمودار
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4]
y = [10, 20, 25, 30]


plt.plot(x, y, marker='o', color='red', linestyle='--')


plt.grid(which='major', color='black', linestyle='-', linewidth=0.8)


plt.minorticks_on()
plt.grid(which='minor', color='gray', linestyle=':', linewidth=0.5)




plt.show()
```
