```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a dataset for balloons
size = 100
std = 0.5
balloons_length = np.random.normal(30,std,size)
balloons_width = np.random.normal(30,std,size)
balloons_weight = np.random.normal(0.1,0,size)
titles = ["length","width","weight"]
balloons_dataset = pd.DataFrame(np.array([balloons_length,balloons_width,balloons_weight]).T,columns=titles)
print("The head of balloons_dataset: \n")
print(balloons_dataset.head())

# Create a dataset for melons
size = 100
std = 0.8
melons_length = np.random.normal(40,std,size)
melons_width = np.random.normal(20,std,size)
melons_weight = np.random.normal(3,std,size)
titles = ["length","width","weight"]
melons_dataset = pd.DataFrame(np.array([melons_length,melons_width,melons_weight]).T,columns=titles)
print("The head of melons_dataset: \n")
print(melons_dataset.head())
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")

ax.scatter(balloons_dataset["length"],balloons_dataset["width"],balloons_dataset["weight"],color="red")
ax.scatter(melons_dataset["length"],melons_dataset["width"],melons_dataset["weight"],color="green")
ax.set_xlabel('length')
ax.set_ylabel('width')
ax.set_zlabel('weight')
ax.legend(["Balloons","Melons"])
plt.savefig("result.jpg")
plt.show()
```
