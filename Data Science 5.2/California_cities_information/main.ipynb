```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("california_cities.csv")
data.head()
```

```python
plt.scatter(data["latd"],
            data["longd"],
            s=data["area_total_km2"],
            alpha=0.5,
            linewidths=0,
            c=np.log10(data["population_total"]),
            cmap="Set1")
plt.colorbar(label="Log10 (population)")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Area and Population of California Cities")
plt.savefig("result.jpg")
plt.show()
```
