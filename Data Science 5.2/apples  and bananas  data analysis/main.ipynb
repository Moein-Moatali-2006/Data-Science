```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 1200
std = 0.5
# Create a dataset for apples
apples_length = np.random.normal(6,std,size)
apples_width = np.random.normal(6,std,size)
columns = ["length","width"]
apples_dataset = pd.DataFrame(np.array([apples_length,apples_width]).T,columns=columns)
print("The head of apples_dataset: \n")
print(apples_dataset.head())

# Create a dataset for bananas
bananas_length = np.random.normal(4,std,size)
bananas_width = np.random.normal(8,std,size)
columns = ["length","width"]
bananas_dataset = pd.DataFrame(np.array([bananas_length,bananas_width]).T,columns=columns)
print("The head of bananas_dataset: \n")
print(bananas_dataset.head())
```

```python
plt.title("Calculation diagram of apples and bananas")
plt.scatter(apples_dataset["length"],apples_dataset["width"],color="red")
plt.scatter(bananas_dataset["length"],bananas_dataset["width"],color="yellow")
plt.legend(["Apples","Bananas"],loc="upper right")
plt.xlabel("Width")
plt.ylabel("length")
plt.savefig("apples_bananas.jpg")
plt.show()
```
